# 🐍 Python Structured Programming: The Basics

## 1️⃣ Input / Output
Python uses `input()` to take user input (always as a string) and `print()` to display output.

```python
# Output
print("Hello, World!")

# Input
name = input("Enter your name: ")
print("Hello, " + name + "!")
```

## 2️⃣ Variables
Variables store data values. No need to declare types explicitly in Python.

```python
# Assigning variables
age = 25        # integer
height = 5.9    # float
name = "Alice"  # string

# Using variables
print(name, "is", age, "years old and", height, "feet tall.")
```

## 3️⃣ Data Types
Python has several basic data types:
- Numbers: int, float
- Strings: str
- Boolean: bool

```python
# Numbers
x = 10        # int
y = 3.14      # float

# String
text = "Python"

# Boolean
is_active = True

print(type(x), type(y), type(text), type(is_active))
```

## 4️⃣ String Formatting
You can insert variables into strings using f-strings, `.format()`, or concatenation.

```python
name = "Alice"
age = 25

# f-string
print(f"My name is {name} and I am {age} years old.")

# .format()
print("My name is {} and I am {} years old.".format(name, age))

# Concatenation
print("My name is " + name + " and I am " + str(age) + " years old.")
```

## 5️⃣ Operators
Python supports:
- Arithmetic: `+`, `-`, `*`, `/`, `//`, `%`, `**`
- Comparison: `==`, `!=`, `>`, `<`, `>=`, `<=`
- Logical: `and`, `or`, `not`

```python
a = 10
b = 3

# Arithmetic
print(a + b, a - b, a * b, a / b, a // b, a % b, a ** b)

# Comparison
print(a > b, a == b, a != b)

# Logical
print((a > 5) and (b < 5))
print(not(a > b))
```

## 6️⃣ Lists
Lists store ordered collections of items.

```python
fruits = ["apple", "banana", "cherry"]

# Access elements
print(fruits[0])  # apple

# Add elements
fruits.append("orange")
print(fruits)

# Remove elements
fruits.remove("banana")
print(fruits)

# Loop through list
for fruit in fruits:
    print(fruit)
```

## 7️⃣ Conditionals
Conditionals allow decision-making: `if`, `elif`, `else`.

```python
age = 18

if age >= 18:
    print("You are an adult.")
elif age >= 13:
    print("You are a teenager.")
else:
    print("You are a child.")
```

## 8️⃣ Loops
Loops help repeat tasks:

### a) `for` loop

```python
for i in range(5):
    print("Iteration:", i)
```

### b) `while` loop

```python
count = 0
while count < 5:
    print("Count:", count)
    count += 1
```

## 9️⃣ Functions
Functions encapsulate reusable code.

```python
# Define a function
def greet(name):
    return f"Hello, {name}!"

# Call function
message = greet("Alice")
print(message)

# Function with default argument
def power(x, y=2):
    return x ** y

print(power(3))    # 3^2 = 9
print(power(2, 3)) # 2^3 = 8
```
