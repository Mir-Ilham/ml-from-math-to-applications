# 📐 Calculus for Machine Learning

## 1️⃣ Differentiation

### 1. Intuition

Differentiation measures how fast a function changes with respect to its input.

If we think of a function as a curve, the derivative represents the slope of the tangent line at a point.

In machine learning, derivatives tell us how changing parameters affects model error.

### 2. Formal Definition
The derivative of a function f(x) is defined as the limit of the rate of change.
$$
f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}
$$

### 3. Simple Example
Consider the function:
$$
f(x) = x^2
$$
Its derivative is:
$$
f'(x) = 2x
$$
At $x=3$, the slope is:
$$
f'(3) = 6
$$

### 4. Visualization
We can visualize a derivative as the slope of the tangent line touching the curve at a specific point.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)
y = x**2
# Given x1 = 3, y1 = 9, m = 6
tangent = 6*x - 9

plt.plot(x,y,label="f(x)=x^2")
plt.plot(x,tangent,label="f'(3)")
plt.axvline(3,color='gray',linestyle='--')

plt.legend()
plt.title("Function Example")
plt.show()
```

### 5. ML Connection
Derivatives are the core of gradient descent.

Example loss function:
$$
L = (y - \hat{y})^2
$$
The derivative tells us how to adjust model parameters to reduce error.

## 2️⃣ Maxima and Minima

### 1. Intuition
Maxima and minima are the highest or lowest points of a function.
- Maximum → highest point
- Minimum → lowest point

Optimization in machine learning means finding the minimum of a loss function.

### 2. Formal Definition
A point is a local minimum or maximum when the derivative is zero.
$$
f'(x) = 0
$$

### 3. Simple Example
Consider the function:
$$
f(x) = x^2
$$
Derivative:
$$
f'(x) = 2x
$$
Set derivative to zero:
$$
2x = 0
$$
Solution:
$$
x = 0
$$
This is the minimum point.

### 4. Visualization
The minimum occurs at the bottom of the parabola.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)
y = x**2

plt.plot(x,y)
plt.scatter(0,0,color='red')

plt.title("Minimum of f(x)=x^2")
plt.show()
```

### 5. ML Connection
Training ML models involves minimizing a loss function.

Example loss:
$$
L(w) = (y - wx)^2
$$
Optimization algorithms like gradient descent find the minimum of this function.

## 3️⃣ Chain Rule

### 1. Intuition
The chain rule is used when one function is inside another function.

If a variable changes through multiple transformations, the chain rule tells us how the final output changes.

### 2. Formal Definition

If
$$
y = f(g(x))
$$

Then the derivative is:
$$
\frac{dy}{dx} = \frac{dy}{dg} \cdot \frac{dg}{dx}
$$

### 3. Simple Example
$$
y = (x^2 + 1)^3
$$
Step 1:
$$
u = x^2 + 1
$$
Step 2:
$$
y = u^3
$$
Derivative:
$$
\frac{dy}{dx} = 3(x^2+1)^2 \cdot 2x
$$

### 4. ML Connection
Chain rule is the foundation of backpropagation.

Neural networks compute:
$$
Loss→Layer3→Layer2→Layer1
$$
Gradients are computed using chain rule through each layer.

## 4️⃣ Partial Derivatives

### 1. Intuition
Partial derivatives measure how a function changes with respect to one variable while keeping others constant.

Many ML models depend on multiple variables.

### 2. Formal Definition
If a function depends on multiple variables:
$$
f(x,y)
$$
Partial derivative with respect to x:
$$
\frac{\partial f}{\partial x}
$$
Partial derivative with respect to y:
$$
\frac{\partial f}{\partial y}
$$

### 3. Simple Example
Let
$$
f(x,y) = x^2 + y^2
$$
Partial derivative with respect to x:
$$
\frac{\partial f}{\partial x} = 2x
$$
Partial derivative with respect to y:
$$
\frac{\partial f}{\partial y} = 2y
$$

### 4. ML Connection
Machine learning models depend on many parameters:
$$
L(w_1, w_2, ..., w_n)
$$
Gradient descent computes:
$$
\frac{\partial L}{\partial w_1}, \frac{\partial L}{\partial w_2}, ...
$$
These partial derivatives form the gradient vector used to update model parameters.
