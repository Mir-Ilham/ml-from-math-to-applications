# 🎲 Probability for Machine Learning

## 1️⃣ Probability Basics

### 1. Intuition
Probability measures how likely an event is to occur.

The value of probability always lies between:
- 0 → impossible event
- 1 → certain event

Example: Probability of getting heads when flipping a fair coin = 0.5

## 2. Formal Definition
If an event 𝐴 occurs in a sample space S:
$$
P(A) = \frac{\text{Number of favorable outcomes}}{\text{Total number of outcomes}}
$$

### 3. Simple Example
Rolling a fair die.

Event 𝐴: rolling a 4
$$
P(A) = \frac{1}{6}
$$

### 4. Basic Probability Rules
#### Complement Rule
$$
P(A^c) = 1 - P(A)
$$
#### Addition Rule (Mutually Exclusive)
$$
P(A \cup B) = P(A) + P(B)
$$

```python
import random

# simulate coin flips
flips = [random.choice(["H","T"]) for _ in range(1000)]

prob_heads = flips.count("H") / len(flips)

print(prob_heads)
```

### 5. ML Connection
Probability is fundamental to ML:

Examples:
- Naive Bayes classifiers
- probabilistic graphical models
- Bayesian learning

## 2️⃣ Conditional Probability & Bayes Theorem

### 1. Intuition
Conditional probability measures: probability of an event given that another event has already occurred.

Example: Probability that a patient has a disease given that a test is positive.

### 2. Conditional Probability 
$$
P(A|B) = \frac{P(A \cap B)}{P(B)}
$$

Where:
- P(A∣B) → probability of A given B
- P(A∩B) → probability both occur

### 3. Simple Example
Suppose:
- 30 students like math
- 20 like programming
- 10 like both
$$
P(Math|Programming) = \frac{10}{20} = 0.5
$$

## Bayes Theorem
### 1. Intuition
Bayes theorem reverses conditional probabilities.

It lets us compute: probability of a cause given observed evidence.

Example:
- disease detection
- spam filtering
- classification

### 2. Formal Definition
$$
P(A|B) =
\frac{P(B|A)P(A)}{P(B)}
$$
Where:
- P(A) → prior probability
- P(B∣A) → likelihood
- P(A∣B) → posterior probability

### 3. Simple Example
Suppose:
- Disease prevalence = 1%
- Test accuracy = 99%

Bayes theorem helps compute the true probability of disease after a positive test.

### 4. ML Connection
Bayes theorem powers:
- Naive Bayes classifiers
- Bayesian inference
- probabilistic machine learning

Naive Bayes uses:
$$
P(y|x_1,x_2,...,x_n)
$$

## 3️⃣ Random Variables & Probability Distributions

### 1. Intuition
A random variable is a variable whose value depends on the outcome of a random event.

Examples:
- number of heads in coin flips
- daily rainfall
- stock prices

### 2. Types of Random Variables
#### Discrete
Values are countable.
- number of customers
- dice roll

#### Continuous
Values are real numbers.
- height
- temperature
- time

### 3. Probability Distribution
A probability distribution describes: how probabilities are assigned to values of a random variable.

## 4️⃣ Normal Distribution, Z-Score & Log-Normal Distribution

## Normal Distribution
### 1. Intuition
The normal distribution is the most important distribution in statistics.

It has a bell-shaped curve.

Properties:
- symmetric
- mean = median = mode

### 2. Probability Density Function
$$
f(x) =
\frac{1}{\sigma\sqrt{2\pi}}
e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$
Where:
- 𝜇 = mean
- 𝜎 = standard deviation

### 3.Visualization
The curve follows the 68-95-99.7 rule:
- 68% within 1 standard deviation
- 95% within 2
- 99.7% within 3

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0,1,1000)

plt.hist(data,bins=30)
plt.show()
```

## Z-Score
### 1. Intuition
The Z-score tells how many standard deviations a value is from the mean.

### 2. Formula
$$
z = \frac{x-\mu}{\sigma}
$$

### 3. Example
- mean = 70
- std = 10
- score = 90
$$
z = \frac{90-70}{10} = 2
$$
Meaning: the value is 2 standard deviations above the mean

## Log-Normal Distribution
### 1. Intuition

A variable follows a log-normal distribution if: the logarithm of the variable is normally distributed.

This distribution is skewed to the right.

### 2. Definition
If:
$$
Y = \ln(X)
$$
and 𝑌 follows a normal distribution, then 𝑋 follows a log-normal distribution.

### 3. Example
Log-normal distributions appear in:
- income distribution
- stock prices
- population growth

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.lognormal(mean=0, sigma=1, size=1000)

plt.hist(data,bins=30)
plt.show()
```
