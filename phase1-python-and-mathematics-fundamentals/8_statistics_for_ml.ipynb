# 📊 Statistics for Machine Learning

## 1️⃣ Mean, Median, Mode, Quartiles & Percentiles

## Mean
### 1. Intuition
The mean (average) represents the central value of a dataset.

It answers the question: “If all values were evenly distributed, what value would each point have?”

### 2. Formal Definition
For a dataset: $x1,x2,x3,...,xn$

The mean is:
$$
\mu = \frac{1}{n}\sum_{i=1}^{n}x_i
$$

### 3. Simple Example
Data: $[2, 4, 6, 8]$
$$
\mu = \frac{2+4+6+8}{4} = 5
$$

```python
import numpy as np

data = np.array([2,4,6,8])

print(np.mean(data))
```

### 4. ML Connection
Mean is used in:
- feature normalization
- loss functions
- gradient descent averaging

## Median
### 1. Definition
The median is the middle value of a sorted dataset.

It is robust to outliers.

### 2. Simple Example
Dataset: $[1,3,5,7,100]$

Median = 5

Even sized dataset: $[1,3,5,7]$

Median:
$
\frac{3+5}{2} = 4
$

```python
import numpy as np

data = np.array([1,3,5,7,100])

print(np.median(data))
```

## Mode
### 1. Definition
The mode is the most frequently occurring value in a dataset.

### 2. Simple Example
Dataset: $[1,2,2,3,4]$

Mode = 2

```python
from scipy import stats

data = [1,2,2,3,4]

print(stats.mode(data))
```

## Quartiles
1. Definition
Quartiles divide data into four equal parts.
- Q1 → 25th percentile
- Q2 → median
- Q3 → 75th percentile

A percentile tells us the percentage of data below a value.

e.g. 90th percentile → 90% of values are below it.

### 2. Example
Dataset: $[1,2,3,4,5,6,7,8]$
Quartiles:
$$
Q1 = 2.5
$$
$$
Q2 = 4.5
$$
$$
Q3 = 6.5
$$

```python
import numpy as np

data = np.array([1,2,3,4,5,6,7,8])

print(np.percentile(data,25))
print(np.percentile(data,50))
print(np.percentile(data,75))
```

## 2️⃣ Range & Interquartile Range

## Range
### 1. Intuition
Range measures spread of the data.

Difference between the largest and smallest value.

### 2. Formal Definition
$$
\text{Range} = x_{max} - x_{min}
$$

### 3. Example
Dataset: $[2,4,6,10]$
$$
Range = 10 - 2 = 8
$$

## Interquartile Range (IQR)
### 1. Intuition
The IQR measures spread of the middle 50% of data.

### 2. Formal Definition
$$
IQR = Q3 - Q1
$$

### 3. Example
$$
Q1 = 2
$$
$$
Q3 = 8
$$
$$
IQR = 8 - 2 = 6
$$

### 4. ML Connection
IQR is used for outlier detection.

Common rule:
$$
\text{Outlier if } x < Q1 - 1.5(IQR)
$$

## 3️⃣ Mean Absolute Deviation, Variance, Standard Deviation

## Mean Absolute Deviation
### 1. Definition
Measures the average distance from the mean.

### 2. Formula
$$
MAD = \frac{1}{n}\sum |x_i - \mu|
$$

## Variance
### 1. Definition
Variance measures how spread out the data is.

### 2. Formula
$$
\sigma^2 = \frac{1}{n}\sum (x_i - \mu)^2
$$

## Standard Deviation
### 1. Definition
Standard deviation is the square root of variance.

### 2. Formula
$$
\sigma = \sqrt{\sigma^2}
$$

```python
import numpy as np

data = np.array([2,4,6,8])

print(np.var(data))
print(np.std(data))
```

## 4️⃣ Skewness

### 1. Intuition
Skewness measures asymmetry of a distribution.

Types:
- Positive skew → long right tail
- Negative skew → long left tail
- Zero skew → symmetric distribution

### 2. Formal Definition
$$
\text{Skewness} =
\frac{1}{n}\sum
\left(
\frac{x_i-\mu}{\sigma}
\right)^3
$$

```python
from scipy.stats import skew
import numpy as np

data = np.array([1,2,3,4,10])

print(skew(data))
```

## 5️⃣ Correlation & Regression Analysis

## Correlation
### 1. Intuition
Correlation measures linear relationship between variables.

Values range:
- -1 → perfect negative
- 0 → no correlation
- +1 → perfect positive

### 2. Formal Definition
Pearson correlation coefficient:
$$
r =
\frac{\sum (x_i-\bar{x})(y_i-\bar{y})}
{\sqrt{\sum (x_i-\bar{x})^2}\sqrt{\sum (y_i-\bar{y})^2}}
$$

```python
import numpy as np

x = np.array([1,2,3,4])
y = np.array([2,4,6,8])

print(np.corrcoef(x,y))
```

## Regression Analysis
### 1. Intuition
Regression finds a relationship between variables.

The simplest model is linear regression.

### 2. Formal Definition
Linear regression model:

$$
y = wx + b
$$
Where:
- 𝑤 → slope
- 𝑏 → intercept

### 3. Code demo

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1],[2],[3],[4]])
y = np.array([2,4,6,8])

model = LinearRegression()
model.fit(X,y)

print(model.coef_)
print(model.intercept_)
```

Regression is one of the most fundamental ML algorithms.

Used in:
- price prediction
- forecasting
- trend analysis

Many ML models extend regression ideas:
- logistic regression
- neural networks
- gradient boosting
