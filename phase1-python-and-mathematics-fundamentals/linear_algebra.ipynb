# 📐 Linear Algebra for Machine Learning

## 1️⃣ Vectors and Matrices

### 1. Intuition
Vectors and matrices are used to represent data and transformations.
- A vector is an ordered list of numbers.
- A matrix is a rectangular table of numbers.

In machine learning:
- vectors represent data points
- matrices represent datasets or transformations

### 2. Formal Definition
A vector with n elements is written as:
$$
\mathbf{v} =
\begin{bmatrix}
v_1 \\
v_2 \\
v_3
\end{bmatrix}
$$
A matrix is written as:
$$
A =
\begin{bmatrix}
a_{11} & a_{12} \\
a_{21} & a_{22}
\end{bmatrix}
$$

### 3. Simple Example
Vector:
$$
v =
\begin{bmatrix}
2 \\
3
\end{bmatrix}
$$
Matrix:
$$
A =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
$$

```python
import numpy as np

v = np.array([2,3])
A = np.array([[1,2],
              [3,4]])

print("Vector:", v)
print("Matrix:\n", A)
```

### 4. ML Connection
Vectors represent feature vectors.

Example data point:
$$
x = [height, weight, age]
$$
Datasets are stored as matrices:
$$
X =
\begin{bmatrix}
x_1 \\

x_3
\end{bmatrix}
$$

## 2️⃣ Scalar Operations

### 1. Intuition
A scalar is a single number.

Scalar operations multiply or divide every element in a vector or matrix.

### 2. Formal Definition
If c is a scalar and v is a vector:
$$
c \cdot v =
\begin{bmatrix}
cv_1 \\
cv_2
\end{bmatrix}
$$

### 3. Simple Example
$$
3
\begin{bmatrix}
1 \\
2
\end{bmatrix}
=
\begin{bmatrix}
3 \\
6
\end{bmatrix}
$$

```python
import numpy as np

v = np.array([1,2])
print(3*v)
```

## 3️⃣ Dot Product

### 1. Intuition
The dot product measures how similar two vectors are.
- Large positive value → vectors point in similar direction
- Zero → vectors are perpendicular

### 2. Formal Definition
For two vectors:
$$
a = (a_1,a_2,...,a_n)
$$
$$
b = (b_1,b_2,...,b_n)
$$
Dot product:
$$
a \cdot b = a_1b_1 + a_2b_2 + ... + a_nb_n
$$

### 3. Simple Example
$$
[1,2] \cdot [3,4] = 1\times3 + 2\times4 = 11
$$

```python
import numpy as np

a = np.array([1,2])
b = np.array([3,4])

print(np.dot(a,b))
```

## 4️⃣ Determinant

### 1. Intuition
The determinant measures how much a matrix scales space.
- determinant = 0 → matrix collapses space
- nonzero determinant → matrix is invertible

### 2. Formal Definition
For a 2×2 matrix:
$$
A =
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$
Determinant:
$$
det(A) = ad - bc
$$

### 3. Simple Example
$$
A =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
$$
$$
det(A) = (1)(4) - (2)(3) = -2
$$

```python
import numpy as np

A = np.array([[1,2],[3,4]])

print(np.linalg.det(A))
```

## 5️⃣ Identity Matrix and Inverse Matrix

### 1. Intuition
The identity matrix acts like the number 1 in matrix multiplication.

Multiplying any matrix by it gives the same matrix.

### 2. Formal Definition
Identity matrix:
$$
I =
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}
$$
Inverse matrix definition:
$$
AA^{-1} = I
$$
Given,
$$
A =
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$
It's inverse will be:
$$
A^{-1} = \frac{1}{det(A)}
\begin{bmatrix}
d & -b \\
-c & a
\end{bmatrix}
$$

### 3. Simple Example
$$
A =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
$$
$$
A^{-1} = \frac{1}{det(A)}
\begin{bmatrix}
4 & -2 \\
-3 & 1
\end{bmatrix}
$$

```python
import numpy as np

A = np.array([[1,2],[3,4]])

print(np.linalg.inv(A))
```

## 6️⃣ Eigenvalues and Eigenvectors

### 1. Intuition
Eigenvectors are directions that do not change direction under a matrix transformation.

They only get scaled.

### 2. Formal Definition
If A is a matrix and v is a vector:
$$
Av = \lambda v
$$
Where:
- 𝑣 = eigenvector
- 𝜆 = eigenvalue

### 3. Simple Example
For matrix:
$$
A =
\begin{bmatrix}
2 & 0 \\
0 & 3
\end{bmatrix}
$$
Eigenvalues:
$$
\lambda_1 = 2, \quad \lambda_2 = 3
$$

```python
import numpy as np

A = np.array([[2,0],
              [0,3]])

values, vectors = np.linalg.eig(A)

print("Eigenvalues:", values)
print("Eigenvectors:\n", vectors)
```

### 4. ML Connection
Eigenvectors are extremely important in ML.
Used in: **Principal Component Analysis (PCA)**

PCA finds the directions of maximum variance.

Covariance matrix:
$$
C = \frac{1}{n}X^TX
$$
Eigenvectors of this matrix give the principal components.
