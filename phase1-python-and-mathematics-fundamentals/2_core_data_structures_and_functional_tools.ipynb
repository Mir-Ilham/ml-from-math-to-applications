# 🐍 Python Data Structures & Functional Tools
This section covers lists, sets, dictionaries, tuples, iterables, list comprehensions, `map`/`filter`, and lambda functions, with explanations and ready-to-run code snippets for Jupyter.

## 1️⃣ Lists, Sets, Dictionaries & Tuples

### Lists
- Ordered, mutable collections.

```python
# List
fruits = ["apple", "banana", "cherry"]
print(fruits)

# Add / remove
fruits.append("orange")
fruits.remove("banana")

# Access elements
print(fruits[0])

# Iterate
for fruit in fruits:
    print(fruit)
```

### Sets
- Unordered, unique elements.

```python
numbers = {1, 2, 3, 3, 4}  # duplicate removed automatically
print(numbers)

# Add / remove
numbers.add(5)
numbers.remove(2)

# Membership test
print(3 in numbers)
```

```python
st = set()

st.add('Apple')
st.add('Orange')
st.add('Apple')

print(st)
```

### Dictionaries
- Key-value pairs, unordered (Python 3.7+ preserves insertion order).

```python
student = {"name": "Alice", "age": 25, "grade": "A"}

# Access
print(student["name"])

# Add / update
student["age"] = 26
student["city"] = "New York"

# Iterate
for key, value in student.items():
    print(key, ":", value)
```

### Tuples
- Ordered, immutable sequences.

```python
point = (10, 20)
print(point)

# Access
print(point[0])

# Unpacking
x, y = point
print("x =", x, "y =", y)
```

## 2️⃣ Iterables and Iterators
- **Iterable**: An object you can loop over (`list`, `tuple`, `dict`, `set`).
- **Iterator**: Object with `__next__()` method; remembers position.

```python
# Iterable
nums = [1, 2, 3]
for n in nums:
    print(n)

# Iterator
it = iter(nums)
print(next(it))
print(next(it))
```

## 3️⃣ List Comprehensions
- Concise way to create lists.

```python
# Normal way
squares = []
for i in range(5):
    squares.append(i**2)
print(squares)

# List comprehension
squares = [i**2 for i in range(5)]
print(squares)

# With condition
even_squares = [i**2 for i in range(10) if i % 2 == 0]
print(even_squares)
```

## 4️⃣ `map()` and `filter()`
- `map(func, iterable)` → applies function to each element.
- `filter(func, iterable)` → keeps elements where func(element) is True.

```python
nums = [1, 2, 3, 4, 5]

# map
squared = list(map(lambda x: x**2, nums))
print("Squared:", squared)

# filter
even = list(filter(lambda x: x % 2 == 0, nums))
print("Even:", even)
```

## 5️⃣ Lambda Functions
- Anonymous one-line functions.

```python
# Normal function
def square(x):
    return x**2
print(square(4))

# Lambda
square = lambda x: x**2
print(square(4))

# Lambda with map
nums = [1, 2, 3]
print(list(map(lambda x: x+1, nums)))
```
