## 1️⃣ Linear and Quadratic Equations

### 1. Intuition
An equation describes a relationship between variables.
- Linear equation → represents a straight line
- Quadratic equation → represents a parabolic curve

These equations describe many real-world relationships.

### 2. Formal Definition

#### Linear Equation
A linear equation in one variable has the form:
$$
ax+b=0
$$
where $𝑎≠0.$

Solution:
$$
x = -\frac{b}{a}
$$

#### Quadratic Equation
A quadratic equation has the form:
$$
ax^2+bx+c=0
$$

Solutions are given by the quadratic formula:
$$
x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$

### 3. Simple Example
#### Linear
$$
2x+4=0
$$
$$
x=−2
$$
#### Quadratic
$$
x^2−5x+6=0
$$
$$
(x−2)(x−3)=0
$$
Solutions:
$$
x=2,x=3
$$

### 4. Visualization
- Linear equation → straight line
- Quadratic equation → parabola

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)

linear = 2*x + 4
quadratic = x**2 - 5*x + 6

plt.plot(x, linear, label="2x + 4")
plt.plot(x, quadratic, label="x^2 - 5x + 6")

plt.axhline(0)
plt.axvline(0)
plt.legend()
plt.title("Linear vs Quadratic Equation")
plt.show()
```

### 5. ML Connection
Linear equations appear in:
- Linear Regression
$$
y=wx+b
$$

Quadratic functions appear in:
- Loss functions
- Optimization landscapes
Example: Mean Squared Error

$$
L=(y−\hat{y})^2
$$

## 2️⃣ Functions

### 1. Intuition
A function maps an input to exactly one output.

Think of it as a machine:

$$
input→function→output
$$

### 2. Formal Definition
A function is defined as:
$$
f:X→Y
$$
where each element of $X$ maps to exactly one element of $Y$.

### 3. Simple Example
$$
f(x)=x^2
$$

```python
def f(x: int):
    return x**2
```

```python
x = 3
f(3)
```

### 4. ML Connection
Machine learning models are essentially functions.

Example:
$$
f(x)=wx+b
$$

Input:
- Features x
Output:
- Prediction y

Neural networks are simply complex functions composed together.

## 3️⃣ Trigonometry Basics

### 1. Intuition
Trigonometry studies relationships between angles and sides of triangles.

It is heavily used in periodic patterns and waves.

### 2. Formal Definition
For a right triangle:
$$
sin(\theta)=\frac{opposite}{hypotenuse}
$$
$$
cos(\theta)=\frac{adjacent}{hypotenuse}
$$
$$
tan(\theta)=\frac{opposite}{adjacent}
$$

### 3. Visualization
Trigonometric functions produce wave-like graphs.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)

plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")

plt.legend()
plt.title("Trigonometric Functions")
plt.show()
```

### 4. ML Connection
Trigonometric functions appear in:
- Fourier transforms
- Signal processing
- Positional encoding in transformers

Example:

Transformer positional encoding uses:
$$
sin(position)
$$
$$
cos(position)
$$

## 4️⃣ Coordinate Geometry Basics

### 1. Intuition

Coordinate geometry represents geometric shapes using coordinates on a plane.

Each point is represented as:
$$
(x, y)
$$

### 2. Formal Definition
A point in 2D space is defined as:
$$
(x, y)
$$
Distance between two points:
$$
d = \sqrt{(x_2-x_1)^2+(y_2-y_1)^2}
$$

### 3. Simple Example
Points:
$$
A = (1,2)
$$
$$
B = (4,6)
$$
Distance:
$$
d = 5
$$

### 4. ML Connection
Coordinate geometry is foundational for:
- Feature space representation
- Distance metrics
- Clustering algorithms

Example distance metric: Euclidean Distance

Used in:
- KNN
- K-Means clustering

## 5️⃣ Graphs

### 1. Intuition
A graph visually represents the relationship between variables.

It helps us understand:
- patterns
- trends
- relationships

### 2. Formal Definition
A graph is the set of points:
$$
(x,f(x))
$$
for a function $f(x)$.

Graphs allow us to see:
- slope
- curvature
- intersections

### 3. ML Connection
Graphs are extremely important in ML:

They help visualize:
- loss curves
- training vs validation performance
- decision boundaries
- data distributions
