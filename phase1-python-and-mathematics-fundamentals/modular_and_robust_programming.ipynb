## 1️⃣ Modules
- Modules are Python files containing functions, classes, or variables.
- Use `import` to include them in your code.

```python
# Importing standard modules
import math
import random

print("Square root of 16:", math.sqrt(16))
print("Random number between 1 and 10:", random.randint(1, 10))

# Import specific functions
from math import factorial
print("5! =", factorial(5))
```

## 2️⃣ File I/O
- Python can read from and write to files.

```python
# Open a file in write mode
with open("example.txt", "w") as file:
    file.write("Hello, World!\n")
    file.write("Python File I/O is easy!")

print("Data written successfully.")
```

```python
with open("example.txt", "r") as file:
    for line in file:
        print(line.strip())
```

## 3️⃣ Exception Handling
- Handle errors gracefully using `try`, `except`, `finally`.

```python
try:
    x = int(input("Enter a number: "))
    y = 10 / x
except ZeroDivisionError:
    print("Error: Division by zero is not allowed!")
except ValueError:
    print("Error: Invalid input. Please enter a number.")
else:
    print("Result:", y)
finally:
    print("Execution completed.")
```

- Raise your own exceptions:

```python
def check_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative!")
    return True

try:
    check_age(-5)
except ValueError as e:
    print("Caught an exception:", e)
```

- Custom Exception Handling

```python
class CustomException(Exception):
    pass

class dobException(CustomException):
    pass

year = int(input("Enter the year of Birth"))
age = 2025 - year

try:
    if 20 <= age and age <= 30:
        print("Valid")
    else:
        raise dobException
except dobException:
    print("The year age is not valid")
```

## 4️⃣ JSON
- JSON is used to store and exchange data in Python. Use the `json` module.

```python
import json

# Python dictionary
data = {"name": "Alice", "age": 25, "skills": ["Python", "ML"]}

# Convert Python object to JSON string
json_str = json.dumps(data)
print("JSON string:", json_str)

# Write JSON to file
with open("data.json", "w") as f:
    json.dump(data, f)

# Read JSON from file
with open("data.json", "r") as f:
    loaded_data = json.load(f)
print("Loaded data:", loaded_data)
```
