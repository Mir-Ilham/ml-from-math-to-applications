# Seaborn Essentials

```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

## 1️⃣ Correlation with Heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in feature selection.

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

## 2️⃣ JointPlot

A joint plot allows to study the relationship between two numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

## 3️⃣ PairPlot

```python
sns.pairplot(df, hue='sex')
```

## 4️⃣ DistPlot

Dist plot helps us to check the distribution of the columns feature

```python
sns.displot(df['tip'], kde=True)
```

### Categorical plots
- box plot
- violin plot
- count plot
- bar plot

## 5️⃣ Count plot

```python
sns.countplot(data=df['day'])
```

## 6️⃣ Bar plot

```python
sns.barplot(y='total_bill', x='smoker', data=df)
```

```python
sns.barplot(x='sex', y='total_bill', data=df)
```

## 7️⃣ Box plot
A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(x=df['smoker'], y=df['total_bill'], hue='smoker', data=df)
```

```python
sns.boxplot(x=df['total_bill'], y=df['day'], hue='smoker', data=df)
```

## 8️⃣ Violin plot

```python
sns.violinplot(x=df['total_bill'], y=df['day'], data=df, hue=df['day'], legend=False)
```
