## 1️⃣ Introduction to Tensors

| Tensor Type         | Dimension |
| ------------------- | --------- |
| Scalar              | 0D        |
| Vector              | 1D        |
| Matrix              | 2D        |
| Higher-order Tensor | 3D+       |

NumPy arrays can represent all of these.

```python
scalar = np.array(5)

vector = np.array([1,2,3])

matrix = np.array([[1,2],
                   [3,4]])

tensor3d = np.array([
    [[1,2],[3,4]],
    [[5,6],[7,8]]
])

print(matrix.shape)
```

| Type    |

| int64   |

| float64 |

```python
a = np.array([1,2,3], dtype=np.float32)

print(a.dtype)
```

## 4️⃣ Scalar Operations
Operations between array and scalar apply to every element.

```python
a = np.array([1,2,3])

print(a + 10)
print(a * 2)
print(a / 2)
```

## 5️⃣ Element-wise Operations
Operations between arrays occur element by element.

Arrays must have same shape or broadcastable shapes.

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
print(a * b)
print(a / b)
```

## 6️⃣ Special Functions
NumPy provides many mathematical functions.
| Function | Description |
| -------- | ----------- |
| `sqrt()` | Square root |
| `exp()`  | Exponential |
| `log()`  | Natural log |
| `sin()`  | Sine        |
| `cos()`  | Cosine      |

```python
a = np.array([1,4,9])

print(np.sqrt(a))

print(np.exp(a))

print(np.log(a))
```

## 7️⃣Comparison Operations
Comparison operations return boolean arrays.

```python
a = np.array([1,5,3])

print(a > 2)

print(a == 3)
```

### Boolean indexing / Conditional filtering

```python
a[a > 2]
```

```python
a[a <= 3]
```

## 8️⃣ Reduction Operations
Reduction operations reduce an array to a single value.
| Function |
| -------- |
| `sum()`  |
| `mean()` |
| `max()`  |
| `min()`  |
| `std()`  |

```python
a = np.array([1,2,3,4])

print(np.sum(a))
print(np.mean(a))
print(np.max(a))
print(np.std(a))
```

```python
# Axis based reduction
matrix = np.array([[1,2],
                   [3,4]])

print(np.sum(matrix, axis=0))
print(np.sum(matrix, axis=1))
```

## 9️⃣ Matrix Operations
NumPy supports matrix algebra operations.

Matrix multiplication uses `@` or `np.dot`.

### Matrix multiplication

```python
A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])

C = A @ B

print(C)
```

## 🔟 Linear Algebra Paradigm vs Looping Paradigm
Instead of looping manually, use vectorized operations.

### Looping (slow)

```python
result = []

for i in range(len(a)):
    result.append(a[i] * 2)

result
```

### Vectorized
Advantages:
- Faster
- Cleaner code
- Uses optimized C backend

## 1️⃣1️⃣ Copying and Reshaping Arrays

### Reshape

```python
a = np.arange(6)

b = a.reshape(2,3)

print(b)
```

```python
b.flatten()
```

### Copy

```python
a = np.array([1,2,3])

b = a.copy()

b[0] = 100

print(a)
```

## 1️⃣2️⃣ In-place Operations
In-place operations modify the original array instead of creating a new one.

```python
a = np.array([1,2,3])

a += 5

print(a)
```
