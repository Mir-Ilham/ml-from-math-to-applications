```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=000fc548033af90caafc50d82c455cbd&language=en-US&page=1')
```

```python
temp_df = pd.DataFrame(response.json()['results'])[['id','adult', 'title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
temp_df.head()
```

```python
partial_dfs = []

for i in range(1,429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=000fc548033af90caafc50d82c455cbd&language=en-US&page={}'.format(i))
    partial_df = pd.DataFrame(response.json()['results'])[['id','adult', 'title','overview','release_date','popularity','vote_average','vote_count']]
    partial_dfs.append(partial_df)

df = pd.concat(partial_dfs, ignore_index=True)
```
