## 5️⃣ Dataframe creation methods

Dataframes can be read from various data sources such as:
- CSV files
- Excel files
- Database

However, data in the form of native python objects can also be converted into dataframes. The methods are discussed below.

### CSV and Excel files

```python
# df = pd.read_csv("weather_data.csv")
# df = pd.read_excel("weather_data.xlsx", sheet_name="Sheet1")
```

### Dictionary of lists

```python
weather_data = {
    'day': ['1/1/2017', '1/2/2017', '1/3/2017'],
    'temperature': [32, 35, 28],
    'windspeed': [6, 7, 2],
    'event': ['Rain', 'Sunny', 'Snow']
}

df = pd.DataFrame(data=weather_data)
df
```

### List of dictionaries

```python
weather_data = [
    {'day': '1/1/2017', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2017', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2017', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'},
]

df = pd.DataFrame(weather_data)
df
```

### List of tuples

```python
weather_data = [
    ('1/1/2017', 32, 6, 'Rain'),
    ('1/2/2017', 35, 7, 'Sunny'),
    ('1/3/2017', 28, 2, 'Snow')
]
df = pd.DataFrame(data=weather_data, columns=["day", "temperature", "windspeed", "event"])
df
```

## 7️⃣ Handling NA values

```python
df = pd.read_csv("weather_data_missing.csv", parse_dates=["day"])
df
```

### Fill all NA

```python
new_df = df.fillna(0)
new_df
```

### Fill NA based on Cols

```python
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'no event'
    })

new_df
```

### bfill and ffill

```python
# new_df = df.ffill()
new_df = df.bfill(limit=1)
new_df
```

### Drop NA

```python
# new_df = df.dropna()
# new_df = df.dropna(how="all")
new_df = df.dropna(thresh=2)

new_df
```

### Concat Method

```python
bangladesh_weather = {
    "city": ["dhaka", "chittagong", "sylhet"],
    "temperature": [32, 45, 30],
    "humidity": [80, 60, 78]
}

bdf = pd.DataFrame(bangladesh_weather)

bdf
```

```python
us_weather = {
    "city": ["new york", "chicago", "orlando"],
    "temperature": [23, 14, 35],
    "humidity": [68, 65, 75]
}

usdf = pd.DataFrame(us_weather)

usdf
```

```python
# Stacking dataframes

df = pd.concat([bdf, usdf])
df
```

```python
# Resetting index

df = pd.concat([bdf, usdf], ignore_index=True)
df
```

```python
# Preserving separation using keys

df = pd.concat([bdf, usdf], keys=["bd", "us"])
df
```

```python
df.loc["bd"]
```

```python
# Appending a new column

s = pd.Series(["Humid", "Dry", "Rain"], name="event")
s
```

```python
df = pd.concat([bdf, s], axis=1)
df
```

### Merge Method

```python
df1 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando", "baltimore"],
    "temperature": [21, 14, 35, 32],
})

df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago", "new york", "san francisco"],
    "humidity": [65, 68, 75],
})

df2
```

```python
# Analogous to SQL joins

# df3 = pd.merge(df1, df2, on="city")
df3 = pd.merge(df1, df2, on="city", how="outer", indicator=True)
# df3 = pd.merge(df1, df2, on="city", how="left")
# df3 = pd.merge(df1, df2, on="city", how="right")

df3
```
